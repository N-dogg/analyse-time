```r
#Random walk analysis
```

```r
simulated.random.walk=arima.sim(n=500,list(order=c(0,1,0)))
```

```r
layout(matrix(c(1,1,2,3),2,2,byrow=TRUE))
plot(simulated.random.walk, main='Simulated Time Series for Random Walk')
hist(simulated.random.walk)
qqnorm(simulated.random.walk)
```

```r
random.walk.diff = diff(simulated.random.walk)
plot.ts(random.walk.diff, main = "First order difference of Random Walk")
```

```r
whitenoise.model = arima(random.walk.diff, order=c(0,0,0))
whitenoise.model
```

```r
plot.ts(simulated.random.walk, main = "First order difference of Random Walk")
abline(0,whitenoise.model$coef)
```
