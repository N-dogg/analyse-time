Simulating a process

```r
#White-noise simulation
simulated.white.noise=arima.sim(n=1000,list(order=c(0,0,0)))
```

```r
layout(matrix(c(1,1,2,3),2,2,byrow=TRUE))
plot(simulated.white.noise, main='Simulated Time Series for White Noise')
hist(simulated.white.noise)
qqnorm(simulated.white.noise)
```

```r
#Random walk simulation
simulated.random.walk=arima.sim(n=1000,list(order=c(0,1,0)))
```

```r
layout(matrix(c(1,1,2,3),2,2,byrow=TRUE))
plot(simulated.random.walk, main='Simulated Time Series for Random Walk')
hist(simulated.random.walk)
qqnorm(simulated.random.walk)
```

```r
#Moving Average Simulation
simulated.ma.1 = arima.sim(n=500, list(order = c(0,0,1), ma=0.5))
simulated.ma.3 = arima.sim(n=500, list(order = c(0,0,3), ma=c(0.5,0.3,0.1)))
```

```r
plot.ts(cbind(simulated.ma.1, simulated.ma.3), main="Moving Average Simulation")
```

```r
plot(cbind(acf(simulated.ma.1), acf(simulated.ma.3), main="Moving Average Simulation"))
```
